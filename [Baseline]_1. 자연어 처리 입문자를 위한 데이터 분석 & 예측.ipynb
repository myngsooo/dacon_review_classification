## 3. ëª¨ë¸ í‰ê°€

ëª¨ë¸ì´ ì–¼ë§ˆë‚˜ ì˜ ë¬¸ì œë¥¼ í’€ì—ˆë‚˜ í™•ì¸í•´ë³´ëŠ” ê³¼ì •ì´ í•„ìš”í•©ë‹ˆë‹¤.

ë¨¼ì € ì–´ë–¤ ê¸°ì¤€ìœ¼ë¡œ ì„±ëŠ¥ì„ ì¸¡ì •í•  ê²ƒì¸ì§€ ì •í•´ì•¼ê² ì£ ?

ì´ë¥¼ í‰ê°€ ì§€í‘œ (Metric)ë¼ê³  í•©ë‹ˆë‹¤.  
ì—¬ê¸°ì„  ì´ë²ˆ ëŒ€íšŒì˜ í‰ê°€ ì§€í‘œì¸ ì •í™•ë„(ACCURACY)ë¥¼ ì‚¬ìš©í•˜ê² ìŠµë‹ˆë‹¤.

```python
# ë¨¼ì € ì ìˆ˜ë¥¼ ë©”ê¸°ëŠ” ë°©ë²•ì¸ í‰ê°€ ì§€í‘œ(Metric)ë¥¼ ì •ì˜í•©ë‹ˆë‹¤.
import numpy as np

def ACCURACY(true, pred):   
    score = np.mean(true==pred)
    return score
```

ì´ì œ ì˜ˆì¸¡ê°’ì„ ìƒì„±í•˜ì—¬ ì‹¤ì œ ì •ë‹µê°’ê³¼ ë¹„êµí•´ë³´ê² ìŠµë‹ˆë‹¤.

```python
# val_Xë¥¼ ì´ìš©í•´ ì˜ˆì¸¡ê°’ì„ ìƒì„±í•©ë‹ˆë‹¤.
y_pred = model.predict(val_X)
```

```python
# ëª¨ë¸ì˜ ì˜ˆì¸¡ê³¼ ì‹¤ì œ ì •ë‹µê°’ì„ ë¹„êµí•©ë‹ˆë‹¤.
score = ACCURACY(val_y, y_pred)

print(f"ëª¨ë¸ì˜ ì •í™•ë„ëŠ” {score*100:.2f}% ì…ë‹ˆë‹¤")
```
